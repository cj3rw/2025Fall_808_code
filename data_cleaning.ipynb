# Data cleaning and preprocessing

```python
# import collected data

import pandas as pd

df_airelationships = pd.read_csv('datasets/df_airelationships.csv')
df_myboyfriendisai = pd.read_csv('datasets/df_myboyfriendisai.csv')
df_mygirlfriendisai = pd.read_csv('datasets/df_mygirlfriendisai.csv')
df_soulmateai = pd.read_csv('datasets/df_soulmateai.csv')

subreddits = [df_airelationships, df_myboyfriendisai, df_mygirlfriendisai, df_soulmateai]

print("len(df_airelationships) is ", len(df_airelationships))
print("len(df_myboyfriendisai) is ", len(df_myboyfriendisai))
print("len(df_mygirlfriendisai) is ", len(df_mygirlfriendisai))
print("len(df_soulmateai) is ", len(df_soulmateai))
```

```python
# Create df_all_unique 
# df_all_unique include all submissions, removed duplicate text, not including submission title 

list_text = []
list_time = []
list_author = []
list_upvotes = [] 
list_subreddit = []

for df in subreddits:

    list_text.extend(df["submission_text"])
    list_time.extend(df["submission_created_utc"])
    list_author.extend(df["submission_author"])
    list_upvotes.extend(df["submission_upvotes"]) 
    list_subreddit.extend(df["submission_subreddit"])

    print(list_author)
    print("next df")

df_all = pd.DataFrame({'text': list_text, 
                       'time': list_time, 
                       'author': list_author, 
                       'upvotes': list_upvotes, 
                       'subreddit': list_subreddit})
print(len(df_all))

# remove duplicated text
df_all_unique = df_all.drop_duplicates(subset=['text'])
print(len(df_all_unique))
```

```python
# remove non-English sentences from each text, rows with empty texts will be removed later

import spacy
from langdetect import detect, LangDetectException

nlp = spacy.load("en_core_web_sm")

def keep_english_sentences_spacy(text):
    doc = nlp(text)
    english_sents = []
    for sent in doc.sents:
        try:
            if detect(sent.text) == "en":
                english_sents.append(sent.text)
        except LangDetectException:
            pass
    return " ".join(english_sents)

df_all_unique["text"] = df_all_unique["text"].fillna("").apply(keep_english_sentences_spacy)
```

```python
# clean all text: lstrip, rstrip, lower, strip punc, rows with empty texts will be removed later
import string 
import emoji

def remove_using_emoji(txt):
    return emoji.replace_emoji(txt, '')

df_all_unique["text"] = [str(x) for x in df_all_unique["text"]]
df_all_unique["text_cleaned"] = [str(x).lstrip().rstrip().lower() for x in df_all_unique["text"]]
df_all_unique["text_cleaned"] = [s.translate(str.maketrans('', '', string.punctuation)) for s in df_all_unique["text_cleaned"]]
df_all_unique["text_cleaned"] = df_all_unique["text_cleaned"].apply(remove_using_emoji)

print(df_all_unique["text_cleaned"].head())
```

```python
# label before or after gpt-5 release (Aug 7, 2025). Unable to find official time point, so use Aug 7, 2025, 00:00:00am

from datetime import datetime
import pandas as pd

df_all_unique["time"] = pd.to_datetime(df_all_unique["time"], utc=True)

gpt5_release = pd.Timestamp("2025-08-07 00:00:00", tz="UTC")

out_list = []
for t in df_all_unique["time"]:
    if t < gpt5_release:
        out_list.append("before")
    else:
        out_list.append("after")

df_all_unique["gpt5_release"] = out_list
```

```python
print(len(df_all_unique))
```

```python
# Remove all links 
df_all_unique["text_cleaned"] = df_all_unique["text_cleaned"].astype(str).str.replace(r'https\S+', '', regex=True)
print(len(df_all_unique))
```

```python
# remove expression that starts with ![gif]
df_all_unique["text"] = df_all_unique["text"].astype(str).str.replace(r'!\[gif\]\(.*?\)', '', regex=True)
print(len(df_all_unique))
```

```python
# remove rows with empty text or just punctuation and emoji
# tried many ways to clean empty text
# considered both "text" and "text_cleaned column"
df_all_unique = df_all_unique[~df_all_unique["text"].isna()]                  
df_all_unique = df_all_unique[df_all_unique["text"].notna() & (df_all_unique["text"].str.strip() != "")]
df_all_unique = df_all_unique[df_all_unique["text"].str.strip().str.len() > 0]
df_all_unique = df_all_unique[df_all_unique["text"].str.replace(r'[^\w\s]', '', regex=True).str.strip().str.len() > 0]

df_all_unique = df_all_unique[~df_all_unique["text_cleaned"].isna()]                 
df_all_unique = df_all_unique[df_all_unique["text_cleaned"].notna() & (df_all_unique["text_cleaned"].str.strip() != "")]
df_all_unique = df_all_unique[df_all_unique["text_cleaned"].str.strip().str.len() > 0]
df_all_unique = df_all_unique[df_all_unique["text_cleaned"].str.replace(r'[^\w\s]', '', regex=True).str.strip().str.len() > 0]
print(len(df_all_unique))
```

```python
# find the subject and objects in all sentences in a text
# find number of human names and counts

import spacy
import pandas as pd

nlp = spacy.load("en_core_web_sm")

def extract_info(text):
    doc = nlp(text)
    human_names = []

    for ent in doc.ents:
        if ent.label_ == "PERSON":
            human_names.append(ent.text)

    return human_names

df_all_unique["human_names"] = df_all_unique["text"].fillna("").apply(extract_info)
df_all_unique["human_name_count"] = df_all_unique["human_names"].apply(len)

print(len(df_all_unique))
```
